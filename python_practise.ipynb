```python
count = 0
word = 'Localization'
vowels = ['a', 'e', 'i', 'o', 'u', 'y', 'h']
for letter in word:
  for i in range(len(vowels)):
    if letter == vowels[i]:
      count += 1
      break
print(f"number of vowels in '{word}' = {count}")
```

```python
primes = []
for i in range(21):
  if i > 1:
    for j in range (2, i):
      if i % j == 0:
        break
    else:
      primes.append(i)

print(f"prime numbers in range (1, 10): {primes}")
```

```python
fib = []
num = int(input("enter any number: "))
for i in range(num):
  if i == 0 or i == 1:
    fib.append(i)
  else:
    fib.append(fib [i - 1] + fib [i - 2])
print(fib)
```
